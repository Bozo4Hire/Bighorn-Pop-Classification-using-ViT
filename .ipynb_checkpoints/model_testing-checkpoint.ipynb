```python
bighorn_ViT_stats = pd.read_csv('./training_stats/bighorn_ViT_stats.csv')
bighorn_CNN_stats = pd.read_csv('./training_stats/bighorn_CNN_stats.csv')
bighorn_ResNet50_stats = pd.read_csv('./training_stats/bighorn_ResNet50_stats.csv')
ua_bighorn_ViT_stats = pd.read_csv('./training_stats/ua_bighorn_ViT_stats.csv')
ua_bighorn_CNN_stats = pd.read_csv('./training_stats/ua_bighorn_CNN_stats.csv')
ua_bighorn_ResNet50_stats = pd.read_csv('./training_stats/ua_bighorn_ResNet50_stats.csv')
```

```python
stat_list = [bighorn_ViT_stats, bighorn_CNN_stats, bighorn_ResNet50_stats, \
             ua_bighorn_ViT_stats, ua_bighorn_CNN_stats, ua_bighorn_ResNet50_stats]
stat_titles = ["ViT", "Simple CNN", "ResNet50 CNN", \
              "ViT", "Simple CNN", "ResNet50 CNN"]
```

```python
print(stat_titles)
```

```python
epochs = list(range(1,11))
```

```python
def plot_stat (stat, stat_title):
    plt.figure(figsize=(12, 5))

    plt.subplot(121)
    plt.plot(epochs, stat_list[0][stat], 'r', label=stat_titles[0])
    plt.plot(epochs, stat_list[1][stat], 'k--', label=stat_titles[1])
    plt.plot(epochs, stat_list[2][stat], 'g', label=stat_titles[2])
    plt.title("Augmented Data")

    plt.subplot(122)
    plt.plot(epochs, stat_list[3][stat], 'r', label=stat_titles[3])
    plt.plot(epochs, stat_list[4][stat], 'k--', label=stat_titles[4])
    plt.plot(epochs, stat_list[5][stat], 'g', label=stat_titles[5])
    plt.title("Unaugmented Data")
    
    ax = plt.subplot(121)
    box = ax.get_position()
    ax.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=5)
    
    ax2 = plt.subplot(122)
    box = ax2.get_position()
    ax2.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])
    ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=5)
    
    plt.suptitle(stat_title)
    plt.show()
    plt.savefig('./figures/'+stat+'.png')
```

```python
plot_stat("train_loss", "Train Loss")
```

```python
plot_stat("train_accuracy", "Train Accuracy")
```

```python
plot_stat("test_loss", "Test Loss")
```

```python
plot_stat("test_accuracy", "Test Accuracy")
```
